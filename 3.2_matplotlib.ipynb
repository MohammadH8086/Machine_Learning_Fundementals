```python
import numpy as np,pandas as pd , matplotlib.pyplot as plt
```

```python
years=[1960,1970,1980,1990,2000,2010,2020]
iranPop=[21.19,28.51,38.67,56.23,66.13,74.75,80.29]
turkeyPop=[27.47,34.88,43.98,53.92,63.24,72.33,79.51]
```

```python
plt.plot(years,iranPop,ls='-',lw=.5,marker='.',mew=3)
plt.plot(years,turkeyPop,ls='--',lw=.5,marker='^',mew=3)
plt.xlabel('Years') ,plt.ylabel('Population'),plt.title('Iran VS Turkey')
plt.legend(['Iran','Turkey'],loc=0)
plt.grid()
plt.annotate('Iran in 1990',xytext=(1975,70),xy=(1990,56.23),c='purple',fontsize=10
            ,arrowprops={'facecolor':'red','width':3})

list_y1=[]
list_y2=[]
for x in range(20,81,10):
    list_y1.append(x)
    list_y2.append(f'{x}m')
plt.yticks(list_y1,list_y2)
```

```python
#plt.figure(figsize=(4,8))
plt.subplot(1,2,1)
plt.plot(years,iranPop,c='red')
plt.title('Iran population',c='purple')

plt.subplot(1,2,2)
plt.plot(years,turkeyPop,c='red')
plt.title('Turkey population',c='purple')
```
