```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.3,shuffle=True,stratify=y)
```

```python
#k_neighbor=np.arange(1,30)
train_accuracy=np.empty(30)
test_accuracy=np.empty(30)

for i,k in enumerate(range(1,31)):
    print(f'i: {i} | k: {k} ')
    knn_loop=KNeighborsClassifier(n_neighbors=k)
    knn_loop.fit(x_train,y_train)
    train_accuracy[i]=knn_loop.score(x_train,y_train)
    test_accuracy[i]=knn_loop.score(x_test,y_test)
plt.plot(np.arange(1,31),train_accuracy,label='train accuracy')
plt.plot(np.arange(1,31),test_accuracy,label='test accuracy')
plt.legend()
plt.xlabel('neighbors')
plt.ylabel('accuracy')
plt.xticks(range(1,31))
```
