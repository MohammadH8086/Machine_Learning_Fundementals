```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_breast_cancer
breast_cancer=load_breast_cancer()
x=breast_cancer.data
y=breast_cancer.target
```

```python
param_grid={'n_neighbors':np.arange(1,51)}
knn=KNeighborsClassifier()
knn_cv=GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(x,y)
```

```python
knn_cv.best_estimator_
```

```python
knn_cv.best_index_
```

```python
knn_cv.best_params_
```

```python
knn_cv.best_score_
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
param_random={'max_depth':[None,3,6,9,12],'max_features':randint(1,9),'min_samples_leaf':randint(1,9)}
decision_tree=DecisionTreeClassifier()
decision_tree_cv=RandomizedSearchCV(decision_tree,param_random,cv=10)
decision_tree_cv.fit(x,y)
```

```python
decision_tree_cv.best_params_
```

```python
decision_tree_cv.best_score_ 
```

```python
decision_tree_cv.score(x,y)
```

```python
from sklearn.naive_bayes import GaussianNB
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,stratify=y)
```

```python
pd.Series(y_test).value_counts()
```

```python
gnb=GaussianNB()
y_predict_gnb=gnb.fit(x_train,y_train).predict(x_test)
```

```python
y_predict_gnb
```

```python
gnb.score(x_train,y_train)
```

```python
gnb.score(x_test,y_test)
```

```python
gnb.score(x_test,y_predict_gnb)
```
