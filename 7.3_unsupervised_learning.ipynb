```python
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.cluster import MeanShift
```

```python
iris=load_iris()
x=iris.data
```

```python
msh=MeanShift(bandwidth=None)
msh.fit(x)
y=msh.predict(x)
centers=msh.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,2],c=y)
plt.scatter(centers[:,0],centers[:,2],marker='x',linewidths=5,s=200)
```
